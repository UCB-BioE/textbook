# Compartmental model of iodine transport to the thyroid 

The thyroid can take up and store iodine. A simple compartmental model of what happens when radioactive iodine is accidentally injested is shown below.

![](https://cdn.mathpix.com/cropped/2024_09_09_d18d4cecd0b381a9bd9dg-6.jpg?height=312&width=1202&top_left_y=1568&top_left_x=451)

QUESTION: What processes are left out of the compartmental model? Why would I leave them out?

Our compartmental model assumes the following:
- Once ingested, iodine moves quickly through the digestive system into the central compartment
- The rate of iodine release from the thyroid is slow compared to the rates
of excretion and transport to the thyroid
- The volume of distribution is approximately constant.
- The thyroid doesn't absorb so much iodine that it becomes "full"
- Before the ingestion, the thyroid was free of radioactive iodine.

So, keep in mind that our model is not good for "short times" (on the order of minutes) because it assumes that movement from the mouth to the stomach, and from the stomach to the central compartment, are instantaneous. This probably takes a few minutes. The model is also not good for "long times" (on the order of weeks) during which we'd need to take into account iodine release from the thyroid. Our model is probably good for "medium times", on the order of hours or maybe a few days - which is fine! That's what we plan on using it for.

The rate of change of Iodine-125 in the central compartment would be:
$$
\frac{d c}{d t}=-\left(k_{t}+k_{e}\right) c
$$

Solve as before to get $\mathrm{c}(\mathrm{t})$, the concentration of iodine in the central compartment:
$$
c(t)=c_{0} e^{-\left(k_{t}+k_{e}\right) t}
$$

QUESTION: How do I get from concentration in the central compartment
to amount in the thyroid?
If I is the amount of iodine stored in the thyroid, then:
$$
\frac{d I}{d t}=k_{t} c V_{d}
$$

QUESTION: What is the $V_{d}$ doing in the equation?
$$
\underbrace{\frac{d I}{d t}}_{\frac{\text { amount }}{\text { time }}}=\underbrace{k_{t}}_{\frac{1}{\text { time }}} \underbrace{c}_{\frac{\text { amount }}{\text { volume }}} \underbrace{V_{d}}_{\text {volume }}
$$

Note that we have an equation for $\mathrm{c}(\mathrm{t})$. Plugging it into the equation gives,
$$
\frac{d I}{d t}=k_{t} V_{d} c_{0} e^{-\left(k_{t}+k_{e}\right) t}
$$

Integrating and solving for $\mathrm{I}(\mathrm{t})$ gives,
$$
I(t)=-\frac{k_{t} V_{d} c_{0}}{k_{e}+k_{t}} e^{-\left(k_{t}+k_{e}\right) t}+A
$$

QUESTION: How do we solve for A?
Given that at $t=0, I(t)=0$,
$$
0=-\frac{k_{t} V_{d} c_{0}}{k_{e}+k_{t}}+A
$$

Which gives,
$$
I(t)=\frac{k_{t} V_{d} c_{0}}{k_{e}+k_{t}}\left(1-e^{-\left(k_{t}+k_{e}\right) t}\right)
$$
