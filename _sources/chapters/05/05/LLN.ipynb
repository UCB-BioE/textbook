# Law of Large Numbers
The Law of Large Numbers (LLN) states that as the size of a sample (or the number of trials) increases, the average of the sample ($\bar{X}$) gets closer and closer to the population mean ($\mu$). 

## Population distribution 
Let's say that we have the following population distribution. 

```python
# load packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# read in data 
path_data = '../../../assets/data/'
df = pd.read_excel(path_data + 'event_count_bimodal_distribution.xlsx')
data = df['Event_Count'].to_numpy()

# Plot the histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='blue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Event Count')
plt.xlabel('Event Count')
plt.ylabel('Frequency')
plt.show()
```

# What happens as we take the average of more and more samples? 
We take n=1 sample, calculate the average, and plot that average. We repeat, with larger and larger sample sizes, n=2, n=3, n=4 ... and plot those values. What happens as n gets large? 

```python
# Number of samples to take
sample_sizes = np.linspace(1, 500000, 200, dtype=int)

# Calculate the population mean
population_mean = np.mean(data)

# Lists to store sample sizes and corresponding sample means
sample_means = []

# Calculate sample means for each sample size
for sample_size in sample_sizes:
    sample = np.random.choice(data, sample_size, replace=True)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(0, max(sample_sizes))
ax.set_ylim(np.mean(data) - 1, np.mean(data) + 1)
ax.set_title('Law of Large Numbers: Convergence of Sample Mean')
ax.set_xlabel('Sample Size')
ax.set_ylabel('Sample Mean')
ax.axhline(population_mean, color='red', linestyle='--', label='Population Mean')
ax.plot(sample_sizes, sample_means, lw=2, label='Sample Mean', color='blue')
ax.legend()
plt.show()
```
